```python
DIR='/Users/ravali/Desktop/DataSciencetsf/data'
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
df1 = pd.read_csv(DIR+'/data1.csv', delimiter=',')
df1.replace([np.inf, -np.inf], np.nan)

df1.dropna()
np.nan_to_num(df1)
df1.dtypes
df1['Order_Demand'] = pd.to_numeric(df1['Order_Demand'], errors='coerce')
df1.dtypes

df1['Order_Demand'] = df1['Order_Demand'].fillna(0)
display(df1.isnull().any())
```

```python

from pandas import datetime
from matplotlib import pyplot

 

df1.plot()
pyplot.show()
```

```python
obj_df = df1.select_dtypes(include=['object']).copy()
obj_df.head()
obj_df["Product_Code"] = obj_df["Product_Code"].astype('category')
obj_df.dtypes
obj_df["Product_Code"] = obj_df["Product_Code"].cat.codes
obj_df["Warehouse"] = obj_df["Warehouse"].astype('category')
obj_df.dtypes
obj_df["Warehouse"] = obj_df["Warehouse"].cat.codes
obj_df["Product_Category"] = obj_df["Product_Category"].astype('category')
obj_df.dtypes
obj_df["Product_Category"] = obj_df["Product_Category"].cat.codes
obj_df1=obj_df.drop('Date',axis=1)
#print(obj_df1)
df1=df1.drop(['Product_Code','Warehouse','Product_Category','Date'],axis=1)
display(df1)
display(df1.dtypes)
```

```python
#df1=df1.drop(['Order_Demand,'Product_code','Warehouse'])
df1 = pd.concat([df1, obj_df1], axis=1)
display(df1)
```
